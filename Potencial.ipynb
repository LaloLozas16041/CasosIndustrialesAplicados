```python
# Cargar los datos
#data_path = '/mnt/data/CreditRiskData.csv'
credit_data = pd.read_csv('CreditRiskData.csv')


# Inspecci칩n inicial de los datos
initial_inspection = {
    'head': credit_data.head(),
    'info': credit_data.info(),
    'describe': credit_data.describe()
}

initial_inspection['head'], initial_inspection['describe']
# La informaci칩n del dataframe se imprimir치 directamente desde credit_data.info()
# Los resultados de 'describe' se mostrar치n a continuaci칩n.
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# 1. Verificaci칩n y manejo de valores at칤picos (Aqu칤 asumiremos que no hay valores at칤picos extremos que requieran atenci칩n inmediata).

# 2. Estandarizar las variables num칠ricas
# Identificar variables num칠ricas
numerical_features = credit_data.select_dtypes(include=['int64', 'float64']).drop(columns='GoodCredit').columns

# 3. Codificar las variables categ칩ricas
# Identificar variables categ칩ricas
categorical_features = credit_data.select_dtypes(include=['object']).columns

# Crear el transformador de columnas para aplicar las transformaciones de manera conjunta
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(), categorical_features)
    ])

# Aplicar el preprocesamiento a las caracter칤sticas predictoras
X = credit_data.drop('GoodCredit', axis=1)
y = credit_data['GoodCredit']
X_processed = preprocessor.fit_transform(X)

# 4. Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
# Estas formas nos dar치n una idea del tama침o de los conjuntos de datos procesados.
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, recall_score, f1_score, roc_auc_score

# Inicializar los modelos
models = {
    'LogisticRegression': LogisticRegression(),
    'SVC': SVC(probability=True),
    'DecisionTreeClassifier': DecisionTreeClassifier(),
    'RandomForestClassifier': RandomForestClassifier(),
    'GradientBoostingClassifier': GradientBoostingClassifier()
}

# Entrenar los modelos y evaluarlos
results = {}

for name, model in models.items():
    # Entrenar el modelo
    model.fit(X_train, y_train)
    
    # Predecir en el conjunto de prueba
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]  # probabilidad para AUC
    
    # Calcular las m칠tricas de evaluaci칩n
    accuracy = accuracy_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_pred_proba)
    
    # Guardar los resultados
    results[name] = {
        'Accuracy': accuracy,
        'Recall': recall,
        'F1 Score': f1,
        'AUC': auc
    }

results  # Este diccionario contiene los resultados de rendimiento de todos los modelos.
```
